```python
## import Python libraries and functionality

import pandas as pd
from scipy import stats
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def jump_diffusion(S_0, r, sigma, lambda_star, S_star, T, Delta_t):
    df = pd.DataFrame()
    N= int(T/Delta_t)
    Z = stats.norm(0,1).rvs(N)
    e = stats.uniform(0,1).rvs(1)
    t = np.linspace(0,T,N)
    df.loc[0,'S_t_hat'] = S_0
    df.loc[0,'lambda_t'] = lambda_star* S_star/S_0
    df.loc[0,'N_t']=0
    df.loc[0,'S_t'] = S_0
    df.loc[0,'default_time'] = 0
    for n in range(0,N-1):
        if (df.loc[t[n],'S_t']==0):
            df.loc[t[n+1],'S_t_hat'] = 0
            df.loc[t[n+1], 'lambda_t'] = np.inf
            df.loc[t[n+1], 'N_t'] = 1.0
            df.loc[t[n+1], 'S_t'] = 0
            df.loc[t[n+1], 'default_time'] = n*Delta_t
        else:
            lambda_t = lambda_star*S_star/df.loc[t[n],'S_t']
            theta_t = (r-0.5*sigma*sigma + lambda_t)*Delta_t + sigma * np.sqrt(Delta_t)*Z[n]
            df.loc[t[n+1],'S_t_hat'] = df.loc[t[n],'S_t']*np.exp(theta_t)
            df.loc[t[n+1], 'lambda_t'] = lambda_t
            df.loc[t[n+1], 'N_t'] = (np.exp(-(df.lambda_t.sum()*Delta_t))< e)*1
            df.loc[t[n+1], 'S_t'] = df.loc[t[n+1],'S_t_hat']*(1-df.loc[t[n+1],'N_t'])  
            df.loc[0,'default_time'] = 0
    return df
```

```python
df = pd.DataFrame()
for i in range(100):
    S = jump_diffusion(100, 0.01, 0.25, 0.05, 100, 5, 0.01)
    df['simulation_'+ str(i)] = S.S_t

df
```

```python
final_values = np.zeros(100)
for j in range(0, 100):
    for i in range(0, 500):
        if (df.iloc[i,j] == 0) and (df.iloc[i-1,j] != 0):
            final_values[j] = 50 * np.exp(-0.01 * (0.01*i))
    if final_values[j] == 0:
        final_values[j] = df.iloc[i,j] * np.exp(-0.01 * 5)
        
final_values
```
